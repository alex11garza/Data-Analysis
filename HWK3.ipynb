An alternative to dropping the missing values would be to replace the missing values with the median for that column since the median is not that sensitive to outliers.

```python
# reread the raw data again
df = pd.read_csv ('Life_Expectancy_Data.csv')
```

```python
# replace the missing values in columns 'Life expectancy', 'GDP', and
# 'Schooling' with the median values for these columns

#get the median values of each listed column
median_life_expectancy = df['Life expectancy'].median()
median_gdp = df['GDP'].median()
median_schooling = df['Schooling'].median()

#fill in NA/missing aleus with median_column
#using .fillna(column, inplace=True)
df['Life expectancy'].fillna(median_life_expectancy, inplace=True)
df['GDP'].fillna(median_gdp, inplace=True)
df['Schooling'].fillna(median_schooling, inplace=True)
```

```python
# check that the replacements took place

# Calculate the number of missing values for each specified column
check_replacement = df[['Life expectancy', 'GDP', 'Schooling']].isnull().sum()

print(check_replacement)
```

##### Task 3: One-Hot Encoding
One-Hot Encoding is a process of transforming categorical data into numerical data. Research how this is done in Pandas.

```python
# reread the raw data again
df = pd.read_csv ('Life_Expectancy_Data.csv')
```

The column 'Status' has categorical data. That data classifies countries as either 'Developing' or 'Developed'.

```python
# check the actual categorical values used
df['Status'].unique()
```

```python
# research the function get_dummies()
Status_Encoded = pd.get_dummies(df['Status'])
print(Status_Encoded)
```

```python
# instead of 'True' and 'False' have 1 and 0.
df.replace({'True': 1, 'False': 0}, inplace=True)

# add the two columns 'Developed' and 'Developing' after 'Status' 
df['Developed'] = (df['Status'] == 'Developed').astype(int)
df['Developing'] = (df['Status'] == 'Developing').astype(int)

cols = df.columns.tolist()
status_idx = cols.index('Status')
cols = cols[:status_idx+1] + ['Developed', 'Developing'] + cols[status_idx+1:-2]
df = df[cols]
```

```python
# check that the columns were added correctly
print(df.columns)
print(df[['Status', 'Developed', 'Developing']].head())
```

```python
# print the number of 'Developed' and 'Developing' countries
num_developed = df['Developed'].sum()
num_developing = df['Developing'].sum()

print(f"Number of Developed countries: {num_developed}")
print(f"Number of Developing countries: {num_developing}")

print(' ')
#number of "Developed" and "Developing" countries in the most recent year of the data (2015)
df_2015 = df[df['Year'] == 2015]
num_developed_2015 = df_2015['Developed'].sum()
num_developing_2015 = df_2015['Developing'].sum()
total_countries_2015 = (df_2015['Developing'].sum()) + (df_2015['Developed'].sum())

print(f"Total number of countries in 2015: {total_countries_2015}")
print(f"Number of Developed countries in 2015: {num_developed_2015}")
print(f"Number of Developing countries in 2015: {num_developing_2015}")
```

```python
# print the mean 'Life expectancy' of 'Developed' countries
# Filter the rows where 'Developed' is 1 and calculate the mean of 'Life expectancy'
mean_life_expectancy_developed = df[df['Developed'] == 1]['Life expectancy'].mean()
print(f"Mean Life expectancy of Developed countries: {mean_life_expectancy_developed:.2f}")
```

##### Task 5: Perform Standardization

Standardization is performed to transform data to have a mean of zero  
and standard deviation of 1.

The standardized value is also called the z-score.

x_z = (x_raw - x_mean) / x_std

Use StandardScaler in scikit-learn to perform the standardization.

```python
# reread the raw data again
df = pd.read_csv ('Life_Expectancy_Data.csv')
```

```python
# get the life expectancy values
df['Life expectancy'].values
```

```python
# Standardization transforms values to have a mean of 0 and standard
# deviation of 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Life expectancy'] = scaler.fit_transform(df['Life expectancy'].values.reshape(-1,1))
```

```python
# check the standardized values
df['Life expectancy']
```

```python
# run simple statistics on that column
df['Life expectancy'].describe()
```

```python
# reread the raw data again
df = pd.read_csv ('Life_Expectancy_Data.csv')
```

```python
# standardize the data for 'Life expectancy', 'GDP', and 'Schooling'
# instead of over-writing the raw values create three columns:
# 'life_expectancy_z', 'gdp_z', and 'schooling_z' and
# fill them with standardized data

df['life_expectancy_z'] = (df['Life expectancy'] - df['Life expectancy'].mean()) / df['Life expectancy'].std()
df['gdp_z'] = (df['GDP'] - df['GDP'].mean()) / df['GDP'].std()
df['schooling_z'] = (df['Schooling'] - df['Schooling'].mean()) / df['Schooling'].std()
```

```python
# check that these three columns were created with standardized data
print(df[['Life expectancy', 'life_expectancy_z', 'GDP', 'gdp_z', 'Schooling', 'schooling_z']].head())
```

##### Task 6: Applying Custom Function to Transform Data

```python
# reread the raw data again
df = pd.read_csv ('Life_Expectancy_Data.csv')
```

```python
# asuume that percentage expenditure has increased by 5%
# define a function that increases all elements by a fixed value of 5% (for simplicity sake) 
def percentage_expenditure_update(balance):
    return balance + 5
```

```python
# apply that function to the DataFrame 
df['percentage expenditure'] = df['percentage expenditure'].apply(percentage_expenditure_update)
```

```python
# check that the values in 'percentage expenditure' did change
# Display the first few rows of 'percentage expenditure'
print(df['percentage expenditure'].head(10))

print(df['percentage expenditure'].describe())
```

```python
# create your own function to perform standardization
# def standardize (mean, std, value):

def standardize(mean, std, value):
    
    return (value - mean) / std


mean_example = df['percentage expenditure'].mean()
std_example = df['percentage expenditure'].std()
value_example = df['percentage expenditure'].iloc[0] 

standardized_value = standardize(mean_example, std_example, value_example)
print(f"Original value: {value_example}, Standardized value: {standardized_value}")

```

```python
# apply the function standardize() to the GDP values.
# store the results in a new column called z_gdp

mean_gdp = df['GDP'].mean()
std_gdp = df['GDP'].std()
df['z_gdp'] = df['GDP'].apply(lambda x: standardize(mean_gdp, std_gdp, x))
print(df[['GDP', 'z_gdp']].head())
```

```python
# compare your standardized values with those obtained from scikit-learn
mean_gdp = df['GDP'].mean()
std_gdp = df['GDP'].std()
df['scikit-learn'] = (df['GDP'] - mean_gdp) / std_gdp
comparison = df[['GDP', 'z_gdp', 'scikit-learn']]
print(comparison.head())
```
